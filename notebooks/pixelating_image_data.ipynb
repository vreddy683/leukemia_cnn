# Creating a DataFrame with pixel values from all and hem images

```python
# Imports to create pixel DF
import numpy as np
from PIL import Image
import cv2
import os
import pandas as pd
```

```python
# Creating variable from data folder
IMG_DIR = 'drive/MyDrive/datasets/C-NMC_Leukemia/training_data/fold_0/all'
print(IMG_DIR)
```

```python
# For loop to turn pixel values into csv (Code modified from https://www.kaggle.com/code/linykc/create-pixel-value-csv-from-images)
for img in os.listdir(IMG_DIR):
    img_array = cv2.imread(os.path.join(IMG_DIR,img), cv2.IMREAD_GRAYSCALE)
    
    img_pil = Image.fromarray(img_array)
    img_28x28 = np.array(img_pil.resize((28, 28), Image.ANTIALIAS))

    img_array = (img_28x28.flatten())

    img_array  = img_array.reshape(-1,1).T

    with open('all_pixel.csv', 'ab') as f:

        np.savetxt(f, img_array, delimiter=",")
```

```python
# Reading in csv
all = pd.read_csv('/content/all_pixel.csv')
```

```python
# Renaming columns as pixel numbers
all.columns = [ 'pixel' + str(i + 1) for i in range(len(all.columns)) ]
all['label'] = 'positive'
all.head()
```

```python
# Creating variable from data folder
IMG_DIR = 'drive/MyDrive/datasets/C-NMC_Leukemia/training_data/fold_0/hem'
print(IMG_DIR)
```

```python
# For loop to turn pixel values into csv
for img in os.listdir(IMG_DIR):
    img_array = cv2.imread(os.path.join(IMG_DIR,img), cv2.IMREAD_GRAYSCALE)
    
    img_pil = Image.fromarray(img_array)
    img_28x28 = np.array(img_pil.resize((28, 28), Image.ANTIALIAS))

    img_array = (img_28x28.flatten())

    img_array  = img_array.reshape(-1,1).T

    with open('hem_pixel.csv', 'ab') as f:

        np.savetxt(f, img_array, delimiter=",")
```

```python
# Reading in csv
hem = pd.read_csv('/content/hem_pixel.csv')
```

```python
# Renaming columns as pixel numbers
hem.columns = [ 'pixel' + str(i + 1) for i in range(len(hem.columns)) ]
hem['label'] = 'negative'
hem.head()
```

```python
# Concatentaing hem and all together
frames = [all, hem]

result = pd.concat(frames)
result.head()
```
